```python
# mfg employment totals starting in 2009m1.
mfg_emp = pd.read_excel('mfg/employment_tot.xlsx', skiprows=12, usecols=['Year', 'Period', 'Value'])

# dropping 2019 values
mfg_emp = mfg_emp[:-3]
```

```python
# creating a datetime index for range of data.
dateind = pd.date_range(start='01/01/2009', end='01/01/2019', freq='M')

# subsetting columns and setting with dateind as index.
mfg_emp['dateind'] = dateind
mfg_emp = mfg_emp[['dateind', 'Value']]
mfg_emp.set_index(['dateind'], inplace=True)
```

```python
# taking initial look at the data. Financial crisis observations may cause issues in regression.
mfg_emp.plot(y='Value')
```

```python
# adding percent change column to view the month to month change.
# using bbfill method so that 2019m1 has same as m2 instead of filling with 0.
mfg_emp['pct change'] = mfg_emp['Value'].pct_change().fillna(method='bfill')
```

```python
# viewing the percent changes. Possibly start data in 2010m1 and remove 2009?
mfg_emp.plot(y='pct change')
```

```python
# view of 2010m1 to 2018m12
mfg_emp.loc['2010-01-31':].plot(y='pct change')
```
