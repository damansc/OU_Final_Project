```python
import pandas as pd
from pandas import datetime
from pandas import DataFrame
from statsmodels.tsa.arima_model import ARIMA
from matplotlib import pyplot
import os
from sklearn.metrics import mean_squared_error
from pandas.plotting import autocorrelation_plot

os.chdir('../data')
```

```python
# data
sep = pd.read_excel('mfg\seperations.xlsx', skiprows=11, usecols=['Value'])
jo = pd.read_excel('mfg\job_openings.xlsx', skiprows=11, usecols=['Value'])

dates = pd.date_range(start='01/01/2009', end='03/01/2019', freq='M')

sep.set_index([dates], inplace=True)
jo.set_index([dates], inplace=True)

print('date index set')
```

```python
# plotting series to visualize the trends.
sep.plot()
jo.plot()
print('Changing data to start in 2010 after the bottom of the recession.')
```

```python
# subsetting the data to start in 2010 as previous periods introduce outlier type data.

sep = sep['01/01/2010':][:]
jo = jo['01/01/2010':][:]

autocorrelation_plot(sep)
pyplot.show()
autocorrelation_plot(jo)
pyplot.show()
```

```python
# 70/30 split for test data and train data

train_end = int(round(len(sep)*.85,0))
test_start = train_end+1

sep_train = sep[:train_end]
sep_test = sep[test_start:]

jo_train = sep[:train_end]
jo_test = sep[test_start:]
```

```python
# fit model
# 5 is lag value for autoregression, 1 is difference order to make stationary, 0 is moving average model.
model = ARIMA(sep_train, order=(5,1,0))
model_fit = model.fit(disp=0)
print(model_fit.summary())

# plot residual errors
residuals = DataFrame(model_fit.resid)
residuals.plot()
pyplot.show()
residuals.plot(kind='kde')
pyplot.show()
print(residuals.describe())
```

```python
test = model_fit.forecast(15)
proj = list(test[0])
sep_test['projections'] = proj
error = mean_squared_error(sep_test.Value, sep_test.projections)
error
```
