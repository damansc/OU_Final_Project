```python
emp = pd.read_excel('48-49/employment_tot.xlsx', skiprows=12, usecols=['Year', 'Period', 'Value'])
```

```python
emp = emp[:-3]
```

```python
newind = pd.date_range(start='01/01/2009', end='01/01/2019', freq='M')
emp['newind'] = newind
emp = emp[['newind', 'Value']]
emp.set_index(['newind'], inplace=True)
```

```python
emp.plot(y='Value')
```

```python
emp['pct change'] = emp['Value'].pct_change().fillna(method='bfill')
```

```python
emp.plot(y='pct change')
```

```python
emp.loc['2010-01-31':].plot(y='pct change')
```
